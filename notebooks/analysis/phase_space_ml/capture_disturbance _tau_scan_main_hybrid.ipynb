```python
# preparations for read box info from the url
import requests
import time

baseUrl = 'http://www.tng-project.org/api/'

def get(path, params=None, max_retries=5, backoff_factor=2):
    # make HTTP GET request to path
    headers = {"api-key":"API KEY"}

    attempt = 0
    while attempt < max_retries:
        try:
            r = requests.get(path, params=params, headers=headers)
            r.raise_for_status()

            if r.headers['content-type'] == 'application/json':
                return r.json()
            
            if 'content-disposition' in r.headers:
                filename = r.headers['content-disposition'].split("filename=")[1]
                with open(filename, 'wb') as f:
                    f.write(r.content)
                return filename
            return r  # fallback
        
        except Exception as e:
            attempt += 1
            wait = backoff_factor ** attempt
            print(f"[Retry {attempt}/{max_retries}] Request failed: {e}. Retrying in {wait}s...")
            time.sleep(wait)

    raise RuntimeError(f"Failed to GET {path} after {max_retries} retries.")
```

```python
# Issue a request to the API root
r = get(baseUrl)

# Print out all the simulation names
names = [sim['name'] for sim in r['simulations']]
# Get the index of TNG300-1
i = names.index('TNG-Cluster')
# Get the info of simulation Illustris-3
sim = get( r['simulations'][i]['url'] )
sim.keys()

# get the snaps info this simulation
snaps = get(sim['snapshots'])
```

```python
def find_redshift_from_lookback_time(tau_gyr):
    """
    Given a look back time, return the redshift
    """

    if tau_gyr >= Planck15.lookback_time(0.01).value:
        tau = tau_gyr * u.Gyr


        def f(z):
            return (Planck15.lookback_time(z) - tau).value  


        result = root_scalar(f, bracket=[0.01, 20], method='brentq')
        
        if result.converged:
            return result.root
        else:
            raise RuntimeError("Root finding failed for tau = {:.2f} Gyr".format(tau_gyr))
    
    else:
        return 0.
```

```python
snap_numbers = np.zeros(len(snaps))
snap_redshifts = np.zeros(len(snaps))
for i, snap in enumerate(snaps):
    snap_numbers[i] = snap['number']
    snap_redshifts[i] = snap['redshift']

def find_nearest_snap(z):
    delta_z = np.abs(snap_redshifts - z)
    target_snap = snap_numbers[np.argmin(delta_z)]
    return int(target_snap)
```

```python
def convert_feats_labels_to_training_set_snapcut_tv(feats_labels_dict, snap_cut, tau):
    feature_vectors = []
    label_scores = []
    label_scores_pre = []
    meta_info = []  

    for halo_id in feats_labels_dict:
        # print(halo_id)
        for snap in feats_labels_dict[halo_id]:
            projections = ['xy', 'blank', 'yz', 'blank', 'xz', 'blank']
            #projections = ['yz', 'blank', 'xz', 'blank', 'xy', 'blank']
            index = np.mod(snap-72,6)
            proj = projections[index]
            if proj == 'blank':
                continue
            if snap <= snap_cut:
                # print(halo_id,'*',snap)
                Redshift = snaps[snap]['redshift'] 
                h_at0 = Planck15.h # unit 100km/[s*Mpc]
                scale = 1 / (h_at0 * (1 + Redshift))
                
                if f'label_score_all_tau{tau}' not in feats_labels_dict[halo_id][snap]:
                    print(f'label_scores_tau{tau} for halo {halo_id} and snap {snap}','vansish')
                    continue
                label = feats_labels_dict[halo_id][snap][f'label_score_all_tau{tau}']
                label_pre = feats_labels_dict[halo_id][snap][f'label_score_pre_tau{tau}']
                #print(snap)
                projections = feats_labels_dict[halo_id][snap]['features']
                

                feats = projections[proj]

                # Convert dict values to a fixed-order feature vector
                vec = [
                    (feats['mean_r_0'] - feats['mean_r_1']) * scale, #0
                    feats['mean_v_0'], #1
                    feats['mean_v_1'], #2
                    np.abs(feats['mean_v_0']-feats['mean_v_1']), #3
                    feats['std_r_0']* scale, #4
                    feats['std_r_1']* scale, #5
                    feats['std_v_0'], #6
                    feats['std_v_1'], #7
                    min(feats['n0'], feats['n1'])/max(feats['n0'], feats['n1']), #8
                    feats['bic_1'], #9
                    feats['bic_2'], #10
                    feats['bic_2'] - feats['bic_1'] - 6*np.log(feats['n0']+ feats['n1']), #11
                    feats['elongation_ratio'],  # 12
                    feats_labels_dict[halo_id][snap]['redshift'], #13
                    feats_labels_dict[halo_id][snap]['mass_ratio'], #14
                    feats['std_r'] * scale, #15
                    feats['std_v'], #16
                ]
                #print(feats['n0']+ feats['n1'])
                feature_vectors.append(vec)
                label_scores.append(label)
                label_scores_pre.append(label_pre)
                meta_info.append((halo_id, snap, proj))

    return np.array(feature_vectors), np.array(label_scores), np.array(label_scores_pre), meta_info
```

```python
def split_test(feature_vectors, label_scores, label_scores_pre, meta_info, test_size, random_state):
    full_halo_ids = np.array([meta_info_fof[0] for meta_info_fof in meta_info])
    unique_halos = np.unique(full_halo_ids)

    halos_train, halos_test = train_test_split(
        unique_halos, test_size=test_size, random_state=random_state
    )

    mask_test = np.isin(full_halo_ids, halos_test)
    mask_train = np.isin(full_halo_ids, halos_train)

    meta_info_train = [meta_info[i] for i in range(len(meta_info)) if mask_train[i]]
    meta_info_test  = [meta_info[i] for i in range(len(meta_info)) if mask_test[i]]

    return (feature_vectors[mask_train], label_scores[mask_train], label_scores_pre[mask_train], meta_info_train,
            feature_vectors[mask_test], label_scores[mask_test], label_scores_pre[mask_test], meta_info_test)
```

```python
def get_feats_label(snapcut, tau):
    with open('/users_path/merger_trace/data/tng_cluster/tng_cluster_products/feats_labels_dict_tngcluster.pkl', 'rb') as f:
        feats_labels_dict = pickle.load(f)

    feature_vectors, label_scores, label_scores_pre, meta_info = convert_feats_labels_to_training_set_snapcut_tv(feats_labels_dict, snapcut, tau)

    return feature_vectors, label_scores, label_scores_pre, meta_info
```

```python
def XGboost_singletau_premerger_allmerger(tau, state = 'pre_merger', random_state = 42):
    if state == 'pre_merger':
        snapcut  = 100
        feature_vectors, label_scores, label_scores_pre, meta_info = get_feats_label(snapcut, tau)
        (feature_vectors_train, label_scores_train, label_scores_pre_train, meta_info_train,
         feature_vectors_test, label_scores_test, label_scores_pre_test, meta_info_test) = split_test(feature_vectors, label_scores, label_scores_pre, meta_info, 0.2, random_state=42)
        label_score_usedfor_train = label_scores_pre_train
        label_score_usedfor_test = label_scores_pre_test
    elif state == 'all_merger':
        cut_z = find_redshift_from_lookback_time(tau)
        snapcut = find_nearest_snap(cut_z)
        feature_vectors, label_scores, label_scores_pre, meta_info = get_feats_label(snapcut, tau)
        (feature_vectors_train, label_scores_train, label_scores_pre_train, meta_info_train,
         feature_vectors_test, label_scores_test, label_scores_pre_test, meta_info_test) = split_test(feature_vectors, label_scores, label_scores_pre, meta_info, 0.2, random_state=42)
        label_score_usedfor_train = label_scores_train
        label_score_usedfor_test = label_scores_test
    
    elif state == 'post_merger':
        cut_z = find_redshift_from_lookback_time(tau)
        snapcut = find_nearest_snap(cut_z)
        feature_vectors, label_scores, label_scores_pre, meta_info = get_feats_label(snapcut, tau)
        (feature_vectors_train, label_scores_train, label_scores_pre_train, meta_info_train,
         feature_vectors_test, label_scores_train, label_scores_pre_test, meta_info_test) = split_test(feature_vectors, label_scores, label_scores_pre, meta_info, 0.2, random_state=2025)
        label_score_usedfor_train = label_scores_train-label_scores_pre_train
        label_score_usedfor_test = label_scores_test-label_scores_pre_test
    
    else:
        print('Please choose the right version of label scores')
    

    # grouping data
    #groups = [f"{hid}_{snap}" for (hid, snap, proj) in np.array(meta_info_train)]
    groups = [hid for (hid, snap, proj) in np.array(meta_info_train)]
    n_samples = len(feature_vectors_train)
    assert len(label_score_usedfor_train) == n_samples == len(groups)

    # split training data and test data
    X = np.arange(len(groups))
    gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=random_state)
    train_idx, test_idx = next(gss.split(X, label_score_usedfor_train, groups=groups))

    X_train, X_test_full = feature_vectors_train[train_idx], np.concatenate([feature_vectors_test, feature_vectors_train[test_idx]])
    y_train, y_test_full = label_score_usedfor_train[train_idx], np.concatenate([label_score_usedfor_test, label_score_usedfor_train[test_idx]])
    X_test_preiso = feature_vectors_test
    y_test_preiso = label_score_usedfor_test
    X_test_split = feature_vectors_train[test_idx]
    y_test_split = label_score_usedfor_train[test_idx]

    group_train = np.array(groups)[train_idx]  # for GridSearchCV

    # set search grid for hyper parameters
    param_grid = {
    'learning_rate': [0.005,0.01,0.05, 0.1, 0.2],
    'max_depth': [1, 2, 4, 6, 8, 10],
    'n_estimators': [100, 200, 300, 500, 1000],
    }

    #initialize model and grid search
    model = XGBRegressor(random_state=random_state)
    cv_splitter = GroupShuffleSplit(n_splits=3, test_size=0.2, random_state=random_state)

    grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='r2',
    cv=cv_splitter,
    n_jobs=-1,
    verbose=1
    )

    grid.fit(X_train, y_train, groups=group_train)

    # predict using best model 

    best_model = grid.best_estimator_
    y_pred_full = best_model.predict(X_test_full)

    mse = mean_squared_error(y_test_full, y_pred_full)
    r2_full = r2_score(y_test_full, y_pred_full)
    print(f"Best Params: {grid.best_params_}")
    print(f"XGBoost MSE: {mse:.4f}")
    print(f"XGBoost R² full: {r2_full:.4f}")

    y_pred_preiso = best_model.predict(X_test_preiso)

    mse = mean_squared_error(y_test_preiso, y_pred_preiso)
    r2_preiso = r2_score(y_test_preiso, y_pred_preiso)
    print(f"Best Params: {grid.best_params_}")
    print(f"XGBoost MSE: {mse:.4f}")
    print(f"XGBoost R² preiso: {r2_preiso:.4f}")

    y_pred_split = best_model.predict(X_test_split)

    mse = mean_squared_error(y_test_split, y_pred_split)
    r2_split = r2_score(y_test_split, y_pred_split)
    print(f"Best Params: {grid.best_params_}")
    print(f"XGBoost MSE: {mse:.4f}")
    print(f"XGBoost R² split: {r2_split:.4f}")

    # visualize data
    plt.figure(figsize=(6, 6))
    plt.scatter(y_test_preiso, y_pred_preiso, alpha=0.5, edgecolors='k', linewidths=0.3)
    plt.plot([0, 5], [0, 5], 'r--', label='Perfect Prediction')
    plt.xlabel('True Merger Score')
    plt.ylabel('Predicted Merger Score')
    plt.title('XGBoost: True vs Predicted')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    y_train_pred = best_model.predict(X_train)
    r2_train = r2_score(y_train, y_train_pred)
    mse_train = mean_squared_error(y_train, y_train_pred)
    print(f"Training R²: {r2_train:.4f}")
    print(f"Training MSE: {mse_train:.4f}")

    meta_info_train_train = [meta_info_train[i] for i in train_idx]
    meta_info_train_test = [meta_info_train[i] for i in test_idx]

    return (mse, r2_preiso, mse_train, r2_train, n_samples, y_test_full, y_pred_full, label_score_usedfor_train, label_score_usedfor_test, meta_info_train_train, meta_info_train_test, meta_info_test
            )
```
